```python
# 1. Write a Function to Check Even or Odd

def check_even_odd(num):
 if num % 2 == 0:
  return "Even"
 else:
  return "Odd"
print(check_even_odd(45))
# The function receives one number as input.
# • num % 2 checks the remainder after division by 2.
# •
# If the remainder is 0, the number is even. Otherwise, it is odd.
# • The function returns the result instead of directly printing it. LihaTech Ko
```

```python
# 2. Write a Function to Find the Largest of Three Numbers
def largest_of_three(a, b, c):
 if a >= b and a >= c:
  return a
 elif b >= a and b >= c:
  return b
 else:
  return c
print("Largest:", largest_of_three(10, 45, 25))

# The function compares a with both b and c.
# •
# If a is greater than or equal to both, a is returned.
# • Otherwise, b is checked in the same way.
# •
# If neither a nor b is largest, c is returned.
```

```python
# 3. Write a Function to Calculate Factorial
def factorial(num):
 fact = 1
 while num > 0:
  fact = fact * num
 num = num - 1
 return fact
print("Factorial:", factorial(5))
# The function starts with fact = 1.
# • The while loop multiplies fact by num repeatedly.
# • After every multiplication, num is decreased by 1.
# • When num becomes 0, the loop ends and the final factorial is returned. LihaTech Kol
```

```python
# 4. Write a Function to Check Prime Number
def is_prime(num):
 if num <= 1:
  return False
 i = 2
 while i < num:
  if num % i == 0:
    return False
 i = i + 1
 return True
print(is_prime(19))
# nmbers less than or equal to
```

```python
# 5. Write a Function to Reverse a String
def reverse_string(text):
 result = ""
 for ch in text:
 result = ch + result
 return result
print(reverse_string("python"))

# # The function starts with an empty string result.
# • For every character in the input text, the character is added before the existing result.
# • This gradually builds the reversed string.
# • Finally, the reversed string is returned.
```

```python
# 6. Write a Function to Count Vowels in a String
def count_vowels(text):
 vowels = "aeiouAEIOU"
 count = 0
 for ch in text:
 if ch in vowels:
 count = count + 1
 return count
print("Vowels:", count_vowels("LihaTech Kolhapur"))


# vowels stores all lowercase and uppercase vowel characters.
# • The function checks each character of the input string.
# •
# If the character is present in vowels, count increases by 1.
# • After the loop, count contains the total number of vowels. L

```

```python
# 7. Write a Function to Check Palindrome String

def is_palindrome(text):
 cleaned = text.lower()
 return cleaned == cleaned[::-1]
print(is_palindrome("Madam"))
```

```python
# 8. Write a Function to Generate Fibonacci Series

def fibonacci(n):
  series =[]
  a,b=0
  count=0

  while count <n:
    series.append(a)
    a,b= b, a+b
    count =count+1
  return series

print(fibonacci(8))

```
