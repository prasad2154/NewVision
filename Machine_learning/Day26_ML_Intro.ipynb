```python
# Prepare data
marks= [[98],[89],[67],[78],[90],[93],[73],[88],[66],[23],[10]]  #input
result=["Pass","Pass","Pass","Pass","Pass","Pass","Pass","Pass","Pass","Fail","Fail"]   #output
```

```python
["Pass" if val[0]>50 else"Fail" for val in marks]
```

```python
# For al Ml Operations we have one of the biggest Library
# Scikit-Larn
# but we import as a sklearn
```

```python
from sklearn.linear_model import LogisticRegression
```

```python

model =LogisticRegression()
model
```

```python
LogisticRegression
```

```python
model
```

```python
# train the model
model.fit(marks,result)
```

```python
model.classes_
```

```python
model.coef_
```

```python
# Test the model
marks,result
```

```python
# Testing
model.predict([[60]])
```

```python
model.predict([[10]])
```

```python
# To check probability
model.predict_proba([[60]])
```
