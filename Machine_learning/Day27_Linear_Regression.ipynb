```python
import pandas as pd
```

```python
path = "/content/Salary_Data.csv"
path
```

```python
# load dataset
sal = pd.read_csv(path)
sal[:3]
```

```python
# Prepare data
x=sal[["YearsExperience"]]  #ml algo always take nputin 2d
y=sal["Salary"]
```

```python
# data analysis
sal["Salary"].values
```

```python
# check relationship of data
import matplotlib.pyplot as plt
plt.scatter(x,y,color="green",marker="*")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.grid()
plt.show()
```

```python
# import algorithm
from sklearn.linear_model import LinearRegression
```

```python
LinearRegression
```

```python
model =LinearRegression()
model
```

```python
model.fit(x,y)
```

```python
# Testing
import warnings
warnings.filterwarnings('ignore')

model.predict([[0]])
```

```python
model.predict([[7.5]])
```

```python
sal.describe().T
```

```python
model.predict([[120]])
```

# **Y=mx+c**

```python
# m-slope
m=model.coef_
m
```

```python
# c-intercept
c=model.intercept_
c
```

```python
m*7.5+c   # model predict the ans with the help of this formula .
```

```python
model.predict([[12]])
```

```python
m*12+c
```

```python
test_data=[[0],[4.5],[6],[11]]
```

```python
y_pred=model.predict(test_data)
y_pred
```

```python
# actual points
plt.scatter(x,y,color="green",marker="*")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
# test line/line of prediction
plt.plot(test_data,y_pred,color="k")
plt.show()
```
