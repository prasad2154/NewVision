```python
# 1. Square a number using lambda.
square = lambda x: x**2
print(square(5))
```

```python
# 2.  Add two numbers.
add = lambda a,b:a+b
print(add(40,20))
```

```python
# 3. Find the larger of two numbers.
maximum = lambda a,b:a if a>b else b
print(maximum(10,20))
```

```python
# 4. Check if a number is even.
even_odd=lambda x:x %2 ==0
print(even_odd(10))
```

```python
# 5. Convert Celsius to Fahrenheit.
Fahrenheit=lambda x:(x*9/5)+32
print(Fahrenheit(30))
```

```python
# 6. Reverse a string.
reverse = lambda d:d[::-1]
print(reverse("Manifastation"))
```

```python
# 7. Find the length of a string.
length=lambda s:len(s)
print(length("Beautifull"))
```

```python
# 8. Multiply four numbers.

multiply = lambda a,b,c,d:a*b*c*d
print(multiply(98,85,79,66))
```

```python
# 9. Check whether a number is positive.
positive = lambda f:"positive" if f>0 else " not positive"
print(positive(98))
```

```python
# 10. Find the last character of a string.
last = lambda x:x[-1]
print(last("Manifastation"))
```

** Map() functions problems** bold text

```python
# 1. Square all numbers.
numbers=[1,2,3,4,5,6]

result = list(map(lambda x:x**2,numbers))

print(result)
```

```python
# 2. cube all numbers.
numbers=[15,25,35,45,55,65]

result = list(map(lambda x:x**3,numbers))

print(result)
```

```python
# 3. Convert names to lowercase
names =["PRASAD","RADHIKA,'SAMIKSHA","RAJ","AASHISH"]
result = list(map(lambda x:x.lower(),names))
print(result)
```

```python
# 4. Convert strings to integers.
numbers =["10","20","30","40","50"]
result = list(map(lambda x:int(x),numbers))
print(result)
```

```python
# 5.Add 15 to every number.

num=[100,200,300]
result = list(map (lambda x:x+15,num))
print(result)
```

```python
# 6. Find lengths of words.
words =["Sammed","Diksha",'Farukh']

result= list(map(lambda x:len(x),words))
print(result)
```

```python
# 7. Convert temperatures to Fahrenheit.
temperatures=[0,45,90]
result=list(map(lambda x:(x*9/5)+32,temperatures))
print(result)
```

```python
#8 Double every number.
numbers =[888,985,654,32,9761,258,741,369]

result=list(map(lambda x:x*2,numbers))
print(result)
```

```python
# 9. Convert words to uppercase.
words = ["cycle","bike","car","scooter","aeroplane"]
result=list(map(lambda x:x.upper(),words))
print(result)
```

```python
# 10.Multiply corresponding elements of two lists.
list1=[1,2,3,4,5]
list2=[6,7,8,9,10]
result=list(map(lambda x,y:x*y,list1,list2))
print(result)
```

**Filter method()**

```python
# 1. Find even numbers.
numbers=[64,22,13,21,45,77.54,65]
result=list(filter(lambda x:x%2==0,numbers))
print(result)
```

```python
# 2.find odd numbers

numbers=[64,22,13,21,45,77.54,65]
result=list(filter(lambda x:x%2!=0,numbers))
print(result)
```

```python
# 3. Find numbers greater than 10.
numbers=[9,64,22,13,21,45,77,7,54,65]

result =list(filter(lambda x:x>10,numbers))
print(result)
```

```python
# 4. Find positive numbers.
numbers=[9,64,22,-13,-21,45,77,7,-54,65]

result =list(filter(lambda x:x>=0,numbers))
print(result)
```

```python
# 5. Find words longer than 5 letters.
words = ["apple","banana","kiwi","orange"]
result = tuple(filter(lambda x:len(x)>5,words))
print(result)
```

```python
# 6. FIlter names starting with 'A'.
names = ["Aarav", "Ananya", "Rohan", "Priya", "Vikram", "Sneha", "Arjun", "Kavya", "Rahul", "Meera"]

result=list(filter(lambda x:x[0]=='A',names))
print(result)
```

```python
# 7. Filter divisible by 3.

num=[10,87,45,65,12,32,75]

result=tuple(filter(lambda x:x%3==0,num))
print(result)
```

```python
# 8.filter empty string
words = ["Python","","CSS","","HTML"]
result = list(filter(lambda x:x!="", words))
print(result)
```

```python
# 9. Filter vowels from a list of characters.
letters = ["a","h","d","i","u"]
result=tuple(filter(lambda x:x in "aeiou",letters))
print(result)
```

```python
# 10. Filter numbers less than 50.
numbers = [20,45,60,45,10]
result=list(filter(lambda h:h<50,numbers))
print(result)
```
