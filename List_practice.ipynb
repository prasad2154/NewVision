### Solutions for Problems 1-10: Basic List Operations

```python
import math
import random
import itertools

# 1. Sum items in a list
def sum_list(items):
    return sum(items)

# 2. Multiply items in a list
def multiply_list(items):
    tot = 1
    for x in items:
        tot *= x
    return tot

# 3. Largest number from a list
def max_num_in_list(list):
    return max(list)

# 4. Smallest number from a list
def min_num_in_list(list):
    return min(list)

# 5. Count strings with length >= 2 and same first/last char
def match_words(words):
    count = 0
    for word in words:
        if len(word) >= 2 and word[0] == word[-1]:
            count += 1
    return count

# 6. Sort list of tuples by last element
def last(n): return n[-1]
def sort_list_last(tuples):
    return sorted(tuples, key=last)

# 7. Remove duplicates
def remove_duplicates(list):
    return list(set(list))

# 8. Check if list is empty
def is_list_empty(list):
    return not list

# 9. Clone or copy a list
def clone_list(li):
    return li[:]

# 10. Words longer than n
def filter_long_words(n, str):
    word_len = []
    txt = str.split(" ")
    for x in txt:
        if len(x) > n:
            word_len.append(x)
    return word_len

print(f"Q1 Sum: {sum_list([1,2,3])}")
print(f"Q5 Count: {match_words(['abc', 'xyz', 'aba', '1221'])}")
print(f"Q6 Sorted: {sort_list_last([(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)])}")
```

```python
# 27. Write a Python program to find the second smallest number in a list.
def second_smallest(numbers):
    a1, a2 = float('inf'), float('inf')
    for x in numbers:
        if x <= a1: a1, a2 = x, a1
        elif x < a2: a2 = x
    return a2

print(second_smallest([1, 2, -8, -2, 0]))
```

```python
# 28. Write a Python program to find the second largest number in a list.
def second_largest(numbers):
    count = 0
    n1 = n2 = float('-inf')
    for x in numbers:
        count += 1
        if x > n1: n1, n2 = x, n1
        elif x > n2 and x != n1: n2 = x
    return n2 if count >= 2 else None

print(second_largest([1, 2, 8, 5, 8, 4]))
```

```python
# 32. Write a Python program to check whether a list contains a sublist.
def is_Sublist(l, s):
    sub_set = False
    if s == []: sub_set = True
    elif s == l: sub_set = True
    elif len(s) > len(l): sub_set = False
    else:
        for i in range(len(l)):
            if l[i] == s[0]:
                n = 1
                while (n < len(s)) and (l[i+n] == s[n]):
                    n += 1
                if n == len(s):
                    sub_set = True
    return sub_set

a = [2, 4, 3, 5, 7]
b = [4, 3]
print(is_Sublist(a, b))
```

```python
# 33. Write a Python program to generate all sublists of a list.
def sub_lists(l):
    base = [[]]
    for i in range(len(l) + 1):
        for j in range(i + 1, len(l) + 1):
            sub = l[i:j]
            base.append(sub)
    return base

l1 = [10, 20, 30]
print(sub_lists(l1))
```

```python
# 10. Write a Python program to find the list of words that are longer than n from a given list of words.
def filter_long_words(n, str):
    word_len = []
    txt = str.split(" ")
    for x in txt:
        if len(x) > n:
            word_len.append(x)
    return word_len

print(filter_long_words(3, "The quick brown fox jumps over the lazy dog"))
```
