```python
#  Positional argument
# these are arg in which sequence order matters

def test(x,y,z):
  print(x,y,z)

test(100,200,400)
test(200,500,700)
```

```python
def info(name,age, sal):
  print('Your name is:',name,'your age is:',age, '& your salary is:',sal  )

info("viki",25,40000)
info("prasad",25,89000)
info(98000,"raj",14)  # when user doesnt follow order then output is nt as per expectations
```

```python
# to solve above problem we have keyword argument
# which doesnot required to follow an order
# bcz it directly refers key=value format

info(sal=98000,name="raj",age=14)
```

```python
# info("viki",21)---we must supply 3 values for 3 arguments if not the it shows error
```

```python
# To avoid missing arguments errors, we can use default arguments.
# Default arguments provide predefined values in the function declaration.
# If the user does not pass a value while calling the function,
# Python automatically uses the default value.

def info(name="Guest", age=25, sal=35000):
    print("Your name is:", name,
          "Your age is:", age,
          "& Your salary is:", sal)

# Function calls
info()                       # Uses all default values
info("Prasad")               # Changes name only
info("Prasad", 22)           # Changes name and age
info("Prasad", 22, 50000)    # Changes all values
```

```python
# Variable length args
# 1.variable length positional arg
# 2.variable length keyword arg

def test(n):
  print(n)
test(10)
```

```python
# 1.variable length positional arg ----to add multiple arguments
# use *
# to solve this issue use
def test(*n):        # * args
  print(n)
test(10,20,30,40)

```

```python
# 2.variable length keyword arg---> **kwargs

def test(**x):
  print(x)

test(num1=45,num2=10,num3=30,num4=88)
```

```python
# when we want to cmbine both we can do so
# but first positional and then keyword

def sample(*args ,**kwargs):
  print(args,kwargs)



sample(10,20,30,40,50,num1=45,num2=10,num3=30,num4=88)

```
