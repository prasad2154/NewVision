**Function with return keyword**

```python
# function will work on specific requirement
# it will generate one result
# this result we need to use an input to the outside of function
# for this purpose we need to return keyword
```

```python
def test(balance):
  if balance <= 0:
    return 'No'
  else:
    return 'Yes'

test(120)
```

```python
# purchase and billing example using return

items = {'rice':56,'chana':30,'wheat':60,'parle_g':10,'milk':35}
bill = []
def purchase(*args):
  for item in args:
    #print(item,items[item])
    bill.append(items[item])
  return bill

def check(total):
  return sum(total)

# calling of purchase
total = purchase('wheat',"milk","rice","chana")
print('Item amounts:',total)

# calling of check
ans=check(total)
print('Total bil amount is Rs:',ans)
```

```python
# calculate % and calculate grade
def calculate_per(marks_obtained,total_marks):
  return(marks_obtained/total_marks)*100


def calculate_grade(per):
  if per>=75:
    return 'Distinction'
  elif per>=65:
    return 'First Class'
  elif per>=55:
    return 'Second class'
  elif per >= 45:
      return 'Pass class'
  else:
    return 'Fail'

calculate_grade(calculate_per(450,500))
```
