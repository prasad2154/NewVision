```python
# 1. Why Exception Handling is Needed
# Concept / Problem Statement: Handle division by zero so that the program does not crash

a = 10
b = 0

try:
  result = a / b
  print(result)
except ZeroDivisionError:
  print("Cannot divide by zero")
```

```python
# 2. Basic try-except Block
# Concept / Problem Statement: Use try-except to handle invalid numeric input.

try:
  age =int(input("Enter your age: "))
  print("Your age is:", age)
except ValueError:
  print("Invalid input. Please enter a valid number")
```

```python
# 3. Handling Multiple Exceptions
# Concept / Problem Statement: Handle both invalid input and division by zero in one program

try:
  a=int(input("enter first number: "))
  b=int(input("enter second number: "))
  result=a/b
  print("result:",result)
except ValueError:
  print("Invalid input. Please enter a valid number")
except ZeroDivisionError:
  print("Cannot divide by zero")
```

```python
# 4. Using except Exception as e
# Concept / Problem Statement: Catch an exception and display the actual error message.

try:
  numbers = (34,50,60)
  print(numbers[5])
except Exception as e:
  print("Error: out of range")
```

```python
# 5. Using else with try-except
# Concept / Problem Statement: Use else to run code only when no exception occurs
try:
  num=int(input("enter a number: "))
except ValueError:
  print("Invalid input. Please enter a valid number")
else:
  print("You entered:", num)
```

```python
# 6. Using finally Block
# Concept / Problem Statement: Use finally to execute code whether an exception occurs or not

try:
  file = open("sample.txt","r")
  print(file.read())
except FileNotFoundError:
  print("File not found")
finally:
  print("Program execution completed")
```

```python
# 7. Raising an Exception Manually
# Concept / Problem Statement: Raise an exception if age is less than 18.


age = 16

try:
  if age<18:
    raise ValueError("Age is less than 18")
except ValueError as e:
  print(e)
```

```python
# 8. Creating a Custom Exception
# Concept / Problem Statement: Create a custom exception for insufficient balance.
class InsufficientBalanceError(Exception):
  pass

balance = 500
withdraw =1000

try:
  if withdraw > balance:
    raise InsufficientBalanceError("Insufficient account balance ")
  else:
    print("Withdraw Successful")
except InsufficientBalanceError:
  print("Insufficient Balance")
```

```python
# # 9. Exception Handling in File Reading
# # Concept / Problem Statement: Safely
# read a file and handle the situation when the file does not exist.

try:
  with open ("file.txt", "r") as file:
    content = file.read()
    print(content)
except FileNotFoundError:
  print("File not found.")
```

```python
# 10. Nested try-except
# Concept / Problem Statement: Use nested try-except to handle errors at different levels

try:
  num=int(input("Enter the a number: "))
  try:
    result=100/num
    print("Result:", result)
  except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
  print("Invalid input. Please enter a valid number")
```

```python
# # 11. Exception Handling with List Access
# Concept / Problem Statement: Handle an error when trying to access an invalid list inde
item = ["python","JAVA",'SQL']
try:
  print(item[3])
except IndexError:
  print("Invalid index")
```

```python
# 12. Using try-except in a Function
# Concept / Problem Statement: Create a function that safely divides two numbers.
def safe_divide(a,b):
  try:
    return a / b
  except ZeroDivisionError:
    return "Cannot divide by zero"

print(safe_divide(10,0))
```
