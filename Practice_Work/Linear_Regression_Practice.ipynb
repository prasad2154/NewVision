```python
import pandas as pd
```

```python
path="/content/mtcars.csv"
path
```

```python
# Load Dataset
car=pd.read_csv(path)
car[:10]
```

```python
X=car[["wt"]]
y=car["mpg"]
```

```python
# data analysis
car["wt"].values
```

```python
# check relationship of data
import matplotlib.pyplot as plt
plt.scatter(X,y,color='orange',marker="*")
plt.xlabel("wt")
plt.ylabel("mpg")
plt.title("Relationship between wt and mpg")
plt.grid()
plt.show()
```

```python
# import algorithm
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model
```

```python
# Train the model
model.fit(X,y)
```

```python
# Testing
import warnings
warnings.filterwarnings('ignore')

model.predict([[3.200]])
```

```python
model.predict([[3]])
```

```python
# m-slope
m=model.coef_
m
```

```python
c=model.intercept_
c
```

```python
m*3.200+c
```

```python
m*7.200+c
```

```python
model.predict([[7.200]])
```

```python
test_data=[[5.2000],[3.870],[1.456]]
```

```python
y_pred=model.predict(test_data)
y_pred
```

```python
# actual plot
plt.scatter(X,y,color='orange',marker="*")
plt.xlabel("wt")
plt.ylabel("mpg")
plt.title("Relationship between wt and mpg")
plt.grid()
# test line of prediction
plt.plot(test_data,y_pred,color='red')
plt.show()
```

# **crop price prediction**

```python
path="/content/Rainfall_vs_Price_dataset.csv"
```

```python
import pandas as pd
```

```python
crop = pd.read_csv(path)
crop[:10]
```

```python
X=crop[["Rainfall_mm"]]
y=crop["Price_per_Quintal"]
```

```python
# Data analysis
crop["Rainfall_mm"].values
```

```python
# check relationship
import matplotlib.pyplot as plt
plt.scatter(X,y,color='orange',marker="*")
plt.xlabel("Rainfall_mm")
plt.ylabel("Price_per_Quintal")
plt.title("Relationship between Rainfall_mm and Price_per_Quintal")
plt.grid()
plt.show()
```

```python
# import algorithm
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
# Train the model
model.fit(X,y)
```

```python
# Testing
import warnings
warnings.filterwarnings('ignore')

model.predict([[280]])
```

```python
model.predict([[550]])
```

```python
# y=mx+x
m=model.coef_
m
```

```python
c=model.intercept_
c
```

```python
m*817+c
```

```python
model.predict([[817]])
```

```python
# check actualpoint

test_data=[[280],[550],[640],[970]]
```

```python
test_data
```

```python
y=model.predict(test_data)
y
```

```python
y_pred_for_plot = y

y_original_for_scatter = crop["Price_per_Quintal"]

plt.scatter(X,y_original_for_scatter,color='orange',marker="o")
plt.xlabel("Rainfall_mm")
plt.ylabel("Price_per_Quintal")
plt.title("Relationship between Rainfall_mm and Price_per_Quintal")

plt.plot(test_data,y_pred_for_plot,color='purple')
plt.show()
# plt.grid()
```
