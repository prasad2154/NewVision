```python
import pandas as pd
```

```python
path="/content/mtcars.csv"
path
```

```python
# Load Dataset
car=pd.read_csv(path)
car[:10]
```

```python
X=car[["wt"]]
y=car["mpg"]
```

```python
# data analysis
car["wt"].values
```

```python
# check relationship of data
import matplotlib.pyplot as plt
plt.scatter(X,y,color='orange',marker="*")
plt.xlabel("wt")
plt.ylabel("mpg")
plt.title("Relationship between wt and mpg")
plt.grid()
plt.show()
```

```python
# import algorithm
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model
```

```python
# Train the model
model.fit(X,y)
```

```python
# Testing
import warnings
warnings.filterwarnings('ignore')

model.predict([[3.200]])
```

```python
model.predict([[3]])
```

```python
# m-slope
m=model.coef_
m
```

```python
c=model.intercept_
c
```

```python
m*3.200+c
```

```python
m*7.200+c
```

```python
model.predict([[7.200]])
```

```python
test_data=[[5.2000],[3.870],[1.456]]
```

```python
y_pred=model.predict(test_data)
y_pred
```

```python
# actual plot
plt.scatter(X,y,color='orange',marker="*")
plt.xlabel("wt")
plt.ylabel("mpg")
plt.title("Relationship between wt and mpg")
plt.grid()
# test line of prediction
plt.plot(test_data,y_pred,color='red')
plt.show()
```
