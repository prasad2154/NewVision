```python
# In pandas we have 2 imp. data structures
# 1. Series - 1D data structure (rows)
# 2. DataFrame -  2D data structure (rows,columns)

# Q. What is the difference between Series and DataFrame???

# A Series is a one-dimensional array that can store a single column of data.

# Like a single column in Excel
# Has index + values
# Can store any data type (int, float, string, etc.)

# A DataFrame is a two-dimensional table made up of multiple Series.

# Like a full Excel table (rows + columns)
# Each column is a Series
# Can store different data types in different columns
```

```python
# How to add a row in df
# A new row can be added by using .loc[].

import pandas as pd


employees = pd.DataFrame({
    "Emp_ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Name": ["Prasad", "Rahul", "Amit", "Sneha", "Neha",
             "Rohan", "Priya", "Karan", "Pooja", "Anjali"],
    "Department": ["IT", "HR", "Sales", "Finance", "IT",
                   "Marketing", "HR", "Sales", "Finance", "IT"],
    "Salary": [50000, 45000, 40000, 55000, 60000,
               48000, 43000, 41000, 57000, 62000]
})

print(employees)
```

```python
# How to add a row in df
# A new row can be added by using .loc[]..
employees.loc[len(employees)]=[111,'Amit',"Sales",50000]
print(employees)
```

```python
# 2. How to Append Two DataFrames

import pandas as pd
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [4, 5, 6], 'B': [7, 8, 9]})
df3 = pd.concat([df1,df2])
print(df3)
```

```python
# 3. How to Concatenate Two or More DataFrames
# Row wise concatenation
# column wise concatenation


# Row wise concatenation
import pandas as pd
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [4, 5, 6], 'B': [7, 8, 9]})
df3 = pd.concat([df1, df2],axis=0)
print(df3)
```

```python
# column wise concatenation
import pandas as pd
df1 = pd.DataFrame({'A': [10, 20, 30], 'B': [40, 50, 60]})
df2 = pd.DataFrame({'C': [40, 50, 60], 'D': [70, 80, 90]})
df3 = pd.concat([df1, df2], axis=1)
print(df3)
```

```python
# 4. How to Perform Merge Operation
# Merge is used to combine DataFrames using a common column


import pandas as pd
student = pd.DataFrame({
    "ID":[101,102,103],
    "Name":["Prasad","Rahul","Amit"]
})
marks = pd.DataFrame({
    "ID":[101,102,103],
    "Marks":[85,90,78]
})
result = pd.merge(student, marks, on="ID")
print(result)


```

```python
# How to Perform Join Operations
import pandas as pd
df1 = pd.DataFrame({
    "Name":["Prasad","Rahul","Amit"]
}, index=[1,2,3])

df2 = pd.DataFrame({
    "Marks":[85,90,78]}, index=[1,2,4])

```

```python
# left join
result = df1.join(df2, how="left")
print(result)
```

```python
# right join
result= df1.join(df2,how="right")
print(result)
```

```python
# outer join
result = df1.join(df2,how="outer")
print(result)
```

```python
# Inner join
result = df1.join(df2,how = "inner")
print(result)
```
