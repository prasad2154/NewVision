## 25 Unique List Comprehension Questions and Solutions

This section provides various exercises to practice and understand list comprehensions in Python.

### Q1: Create a list of the first 10 perfect cubes


```python
# Solution 1
squares = [i**3 for i in range(1, 11)]
display(squares)
```

### Q2: Given a list of words, create a new list containing only words that start with the letter 'a'.

```python
# Solution 2
words = ['apple', 'banana', 'apricot', 'cherry', 'grape', 'avocado']
a_words = [word for word in words if word.startswith('a')]
display(a_words)
```

### Q3: Convert a list of Fahrenheit temperatures to Celsius.

```python
# Solution 3
fahrenheit_temps = [32, 68, 100, 212, 0]
celsius_temps = [(f - 32) * 5/9 for f in fahrenheit_temps]
display(celsius_temps)
```

### Q4: Create a list of even numbers from 1 to 20.

```python
# Solution 4
even_numbers = [i for i in range(1, 21) if i % 2 == 0]
display(even_numbers)
```

### Q5: Flatten a list of lists into a single list.

```python
# Solution 5
list_of_lists = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
flattened_list = [item for sublist in list_of_lists for item in sublist]
display(flattened_list)
```

### Q6: Given a list of numbers, create a new list with only the positive numbers.

```python
# Solution 6
numbers = [-5, -2, 0, 1, 3, 7, -10]
positive_numbers = [num for num in numbers if num > 0]
display(positive_numbers)
```

### Q7: Create a list of all uppercase characters from a given string.

```python
# Solution 7
some_string = "Hello World! Python is FUN."
upper_chars = [char for char in some_string if char.isupper()]
display(upper_chars)
```

### Q8: Given a list of numbers, create a list where each number is doubled if it's even, and tripled if it's odd.

```python
# Solution 8
numbers = [1, 2, 3, 4, 5, 6]
transformed_numbers = [num * 2 if num % 2 == 0 else num * 3 for num in numbers]
display(transformed_numbers)
```

### Q9: Generate a list of tuples (number, square) for numbers from 1 to 5.

```python
# Solution 9
num_square_tuples = [(i, i**2) for i in range(1, 6)]
display(num_square_tuples)
```

### Q10: Create a list of file extensions from a list of filenames.

```python
# Solution 10
filenames = ['document.pdf', 'image.png', 'script.py', 'archive.tar.gz']
extensions = [f.split('.')[-1] for f in filenames]
display(extensions)
```

### Q11: Filter out all non-string elements from a mixed list.

```python
# Solution 11
mixed_list = [1, 'hello', 3.14, 'world', True, None, 'python']
string_list = [item for item in mixed_list if isinstance(item, str)]
display(string_list)
```

### Q12: Create a list of cube roots for numbers from 1 to 10 (rounded to 2 decimal places).

```python
# Solution 12
import math
cube_roots = [round(math.pow(i, 1/3), 2) for i in range(1, 11)]
display(cube_roots)
```

### Q13: Given two lists, create a list of common elements.

```python
# Solution 13
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
common_elements = [item for item in list1 if item in list2]
display(common_elements)
```

### Q14: Generate a list of booleans indicating if a number is divisible by 3 for numbers 1 to 15.

```python
# Solution 14
divisible_by_3 = [i % 3 == 0 for i in range(1, 16)]
display(divisible_by_3)
```

### Q15: Remove all vowels from a given string and return as a list of characters.

```python
# Solution 15
sentence = "This is a sample string."
vowels = "aeiouAEIOU"
consonants_list = [char for char in sentence if char not in vowels]
display(consonants_list)
```

### Q16: Create a list of lengths of words in a sentence.

```python
# Solution 16
sentence = "The quick brown fox jumps over the lazy dog"
word_lengths = [len(word) for word in sentence.split()]
display(word_lengths)
```

### Q17: Generate a list of all possible pairs from two given lists.

```python
# Solution 17
list_a = [1, 2]
list_b = ['a', 'b']
pairs = [(x, y) for x in list_a for y in list_b]
display(pairs)
```

### Q18: Create a list of numbers from 1 to 100 that are divisible by both 3 and 5.

```python
# Solution 18
divisible_by_3_and_5 = [i for i in range(1, 101) if i % 3 == 0 and i % 5 == 0]
display(divisible_by_3_and_5)
```

### Q19: Given a list of dictionaries, extract the values associated with a specific key.

```python
# Solution 19
data = [{'name': 'Alice', 'age': 30}, {'name': 'Bob', 'age': 24}, {'name': 'Charlie', 'age': 35}]
ages = [d['age'] for d in data]
display(ages)
```

### Q20: Create a list of strings where each string is reversed.

```python
# Solution 20
words = ['python', 'java', 'csharp']
reversed_words = [word[::-1] for word in words]
display(reversed_words)
```

### Q21: Filter a list of numbers to include only those greater than their index.

```python
# Solution 21
numbers = [0, 5, 2, 8, 3, 1]
filtered_numbers = [num for index, num in enumerate(numbers) if num > index]
display(filtered_numbers)
```

### Q22: Create a list of characters from a string, excluding spaces.

```python
# Solution 22
sentence = "Python programming is fun"
chars_no_spaces = [char for char in sentence if char != ' ']
display(chars_no_spaces)
```

### Q23: Generate a multiplication table for a given number (e.g., 7) up to 10.

```python
# Solution 23
num = 7
multiplication_table = [f"{num} x {i} = {num * i}" for i in range(1, 11)]
display(multiplication_table)
```

### Q24: From a list of names, create a list of their first initials.

```python
# Solution 24
names = ['Alice Smith', 'Bob Johnson', 'Charlie Brown']
initials = [name.split()[0][0] for name in names]
display(initials)
```

### Q25: Create a list of tuples containing (number, 'even'/'odd') for numbers from 1 to 10.

```python
# Solution 25
even_odd_list = [(i, 'even' if i % 2 == 0 else 'odd') for i in range(1, 11)]
display(even_odd_list)
```
