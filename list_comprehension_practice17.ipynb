## 25 Unique List Comprehension Questions and Solutions

This section provides various exercises to practice and understand list comprehensions in Python.

### Q1: Create a list of the first 10 perfect cubes


```python
# Solution 1
squares = [i**3 for i in range(1, 11)]
display(squares)
```

### Q2: Given a list of words, create a new list containing only words that start with the letter 'a'.

```python
# Solution 2
words = ['apple', 'banana', 'apricot', 'cherry', 'grape', 'avocado']
a_words = [word for word in words if word.startswith('a')]
display(a_words)
```

### Q3: Convert a list of Fahrenheit temperatures to Celsius.

```python
# Solution 3
fahrenheit_temps = [32, 68, 100, 212, 0]
celsius_temps = [(f - 32) * 5/9 for f in fahrenheit_temps]
display(celsius_temps)
```

### Q4: Create a list of even numbers from 1 to 20.

```python
# Solution 4
even_numbers = [i for i in range(1, 21) if i % 2 == 0]
display(even_numbers)
```

### Q5: Flatten a list of lists into a single list.

```python
# Solution 5
list_of_lists = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
flattened_list = [item for sublist in list_of_lists for item in sublist]
display(flattened_list)
```

### Q6: Given a list of numbers, create a new list with only the positive numbers.

```python
# Solution 6
numbers = [-5, -2, 0, 1, 3, 7, -10]
positive_numbers = [num for num in numbers if num > 0]
display(positive_numbers)
```

### Q7: Create a list of all uppercase characters from a given string.

```python
# Solution 7
some_string = "Hello World! Python is FUN."
upper_chars = [char for char in some_string if char.isupper()]
display(upper_chars)
```

### Q8: Given a list of numbers, create a list where each number is doubled if it's even, and tripled if it's odd.

```python
# Solution 8
numbers = [1, 2, 3, 4, 5, 6]
transformed_numbers = [num * 2 if num % 2 == 0 else num * 3 for num in numbers]
display(transformed_numbers)
```

### Q9: Generate a list of tuples (number, square) for numbers from 1 to 5.

```python
# Solution 9
num_square_tuples = [(i, i**2) for i in range(1, 6)]
display(num_square_tuples)
```

### Q10: Create a list of file extensions from a list of filenames.

```python
# Solution 10
filenames = ['document.pdf', 'image.png', 'script.py', 'archive.tar.gz']
extensions = [f.split('.')[-1] for f in filenames]
display(extensions)
```

### Q11: Filter out all non-string elements from a mixed list.

```python
# Solution 11
mixed_list = [1, 'hello', 3.14, 'world', True, None, 'python']
string_list = [item for item in mixed_list if isinstance(item, str)]
display(string_list)
```

### Q12: Create a list of cube roots for numbers from 1 to 10 (rounded to 2 decimal places).

```python
# Solution 12
import math
cube_roots = [round(math.pow(i, 1/3), 2) for i in range(1, 11)]
display(cube_roots)
```

### Q13: Given two lists, create a list of common elements.

```python
# Solution 13
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
common_elements = [item for item in list1 if item in list2]
display(common_elements)
```

### Q15: Remove all vowels from a given string and return as a list of characters.

### Q16: Create a list of lengths of words in a sentence.

### Q17: Generate a list of all possible pairs from two given lists.
