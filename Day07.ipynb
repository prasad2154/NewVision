### `pop()` method

The `pop()` method removes the item with the specified key and returns its value.

```python
print(f"Original dictionary: {x}")
removed_value = x.pop(1) # Remove item with key 1
print(f"Dictionary after pop(1): {x}")
print(f"Removed value: {removed_value}")
```

### `popitem()` method

The `popitem()` method removes the last inserted key-value pair and returns it as a tuple.

```python
print(f"Original dictionary: {x}")
removed_item = x.popitem() # Remove the last item
print(f"Dictionary after popitem(): {x}")
print(f"Removed item: {removed_item}")
```

### `get()` method

The `get()` method returns the value of the item with the specified key. If the key does not exist, it returns `None` by default, or a specified default value, without raising an error.

```python
print(f"Dictionary: {x}")
value1 = x.get(2)
value_non_existent = x.get(10, 'Key not found') # Key 10 does not exist, return 'Key not found'
value_none = x.get(4) # Key 4 exists with value None
print(f"Value for key 2: {value1}")
print(f"Value for non-existent key 10: {value_non_existent}")
print(f"Value for key 4: {value_none}")
```

### `copy()` method

The `copy()` method returns a shallow copy of the dictionary.

```python
print(f"Original dictionary: {x}")
x_copy = x.copy()
print(f"Copied dictionary: {x_copy}")

# Modifying the copy does not affect the original
x_copy[6] = 'New Value'
print(f"Original after copy modification: {x}")
print(f"Copied after modification: {x_copy}")
```

### `fromkeys()` method

The `fromkeys()` method returns a new dictionary with the specified keys and the specified value.

```python
keys = ['a', 'b', 'c']
value = 0
new_dict = dict.fromkeys(keys, value)
print(f"Dictionary from keys with a single value: {new_dict}")

keys_only = ['d', 'e']
new_dict_default = dict.fromkeys(keys_only) # Default value is None
print(f"Dictionary from keys with default None value: {new_dict_default}")
```

### `clear()` method

The `clear()` method removes all the elements from the dictionary.

```python
print(f"Original dictionary: {x}")
x.clear()
print(f"Dictionary after clear(): {x}")
```
