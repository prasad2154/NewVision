```python
# 11. check prime numberss
num = int(input("Enter a number: "))

if num<=1:
  print("not a prime number:")
else:
  for i in range(2,num):
    if num%i==0:
      print("not prime")
      break
    else:
      print("prime")
```

```python
# alternative
import math

num = int(input("Enter a number: "))

if num <= 1:
    print("Not Prime")
else:
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```
