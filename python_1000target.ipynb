```python
# alternative
import math

num = int(input("Enter a number: "))

if num <= 1:
    print("Not Prime")
else:
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```

```python
# 23. Find the remainder without using the % operator.
dividend = int(input("Enter the dividend: "))
divisor = int(input("Enter the divisor: "))

quotient = dividend // divisor
remainder = dividend - (divisor * quotient)

print("The remainder is:", remainder)
```

```python
# 24. Find the quotient without using the / operator.
dividend = int(input("Enter the dividend: "))
divisor = int(input("Enter the divisor: "))

quotient = 0

while dividend >= divisor:
    dividend -= divisor
    quotient += 1

print("The quotient is:", quotient)
```

```python
# 25. Reverse a two-digit number.
num = int(input("Enter a two-digit number: "))
reversed_num = 0

while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num //= 10

print("The reversed number is:", reversed_num)
```

```python
# 27. Find the sum of digits of a number.
num = int(input("Enter a number: "))
sum_of_digits = 0

while num > 0:
    digit = num % 10
    sum_of_digits += digit
    num //= 10

print("The sum of digits is:", sum_of_digits)
```

```python
# 30. Find the first and last digit of a number.
num = int(input("Enter a number: "))
last_digit = num % 10

while num >= 10:
    num //= 10

first_digit = num

print("The first digit is:", first_digit)
print("The last digit is:", last_digit)
```
