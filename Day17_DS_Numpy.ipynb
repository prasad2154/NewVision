```python
x[0]
```

```python
x[4]
```

```python
x[::-1]
```

```python
x[2]
```

```python
print(x[2][3])   # row with element index
```

```python
print(x[5][3])
```

```python
print(x[3][2])
```

```python
# numpy array is mutable in nature
x[3,3]
```

```python
# replace 15 by 50
x[3,3]=50
x
```

```python
x[4,2]=70
x
```

```python
# access more rows
x[:] #[start_index,stop_index]
x[1:4]
```
