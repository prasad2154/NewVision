```python
# all in one example
import numpy as np

b = np.array([56,98,125,3654,987,425,125,42,552,800])

print("Minimum:", b.min())
print("Maximum:", b.max())
print("Mean:", b.mean())
print("Standard Deviation:", b.std())
```
