```python
# Conditional list comprehension
# fetch those names whose len is > 6
# syntax: [operation for var in sequence/iteration if <condition>]
# first element fetched in nm---> check condition len(nm)>6---> perform operation
[nm.title() for nm in nms if len(nm)>6]
```

```python
k = [1, 1, 1, 2, 2, 2, 3, 4, 5]

max_count = 0
most_freq = []

for i in k:
    count = k.count(i)

    if count > max_count:
        max_count = count
        most_freq = [i]

    elif count == max_count and i not in most_freq:
        most_freq.append(i)

print("Most frequent:", most_freq)
print("Frequency:", max_count)
```

```python
from collections import Counter

k = [1, 1, 1, 2, 2, 2, 3, 4, 5]

freq = Counter(k)
max_count = max(freq.values())

most_freq = [num for num, count in freq.items() if count == max_count]

print(most_freq)
```
