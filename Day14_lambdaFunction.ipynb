```python
names=["gayatri","abdul","snehal","jay"]

# convert above in uppercase using map

upper_names_map = map(lambda name: name.upper(), names)
print(list(upper_names_map))
```

```python
text = "apple banana apple cherry banana apple"
words = text.split()

# Count occurrences using a dictionary comprehension
word_counts = {word: words.count(word) for word in set(words)}

print(f"Original text: {text}")
print(f"Word counts: {word_counts}")
```

```python
# filter (function.iterable)
# this is conditional function
# the value for which condition is true ,will get fetched and operation will be performed if required

p=(12,-22,34,-54,0,-32,98,76)
# filter -ve numbers

list(filter(lambda num:num>0,p))

```
