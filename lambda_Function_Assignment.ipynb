```python
# 1. What is a Lambda Function?
# Problem Statement: Create a simple lambda function that adds 10 to a given num

add_ten = lambda x: x + 10
print(add_ten(5))

```

```python
# 2. Lambda Function with Two Arguments
# Problem Statement: Create a lambda function to multiply two numbers

multiply = lambda a,b:a+b
print(multiply(7,9))
```

```python
# 3. Lambda Function for Square of a Number
# Problem Statement: Create a lambda function that returns the square of a given number

square = lambda x:x**2
print(square(75))
```

```python
# 4. Lambda with Conditional Expression
# Problem Statement: Create a lambda function to check whether a number is even or odd.

check_even_odd = lambda x: "even" if x%2==0 else "odd"
print(check_even_odd(19))
print(check_even_odd(90))
```

```python
# 5. Lambda with map()
#  Use lambda with map() to square every number in a list

numbers =[1,4,5,7,8,9,20]

squares =list(map(lambda x:x**2,numbers))
print(squares)
```

```python
# 6. Lambda with filter()
# Problem Statement: Use lambda with filter() to get only even numbers from a list

numbers =[11,42,53,74,85,96,20]

even_numbers =list(filter(lambda x:x%2==0,numbers))
print(even_numbers)
```

```python
# 7. Lambda with sorted()
# Problem Statement: Sort a list of tuples based on the second value using lambda.

students =[('Amit', 85), ('Neha ', 92), ('Chaitanya', 78),]

sorted_students =sorted(students,key=lambda x:x[1])

print(sorted_students)
```

```python
# 8. Lambda with List of Dictionaries
# Problem Statement: Sort employee records based on salary using lambda.

employees = [
    {'name': 'Amit', 'salary': 60000},
    {'name': 'Neha', 'salary': 75000},
    {'name': 'Chaitanya', 'salary': 55000},
]

sorted_employees =sorted(employees,key=lambda x:x['salary'])

print(sorted_employees)
```

```python
# 9. Lambda Function Returning Maximum of Two Numbers
# Problem Statement: Create a lambda function that returns the maximum of two numbers.

maximum = lambda a , b:a if a>b else b
print(maximum (45,98))
```

```python
# 10. Lambda inside Another Function
# Problem Statement: Create a function that returns a lambda function for multiplying a number by a given value

def multiply_by(value):
    return lambda x:x*value

double = multiply_by(2)
triple = multiply_by(3)

print(double(10))
print(triple(10))
```

```python
# 11. Lambda with reduce()
# Problem Statement: Use lambda with reduce() to find the product of all numbers in a list.

from functools import reduce

numbers =[1,2,3,4,5,6,7,8]

product =reduce(lambda x,y:x*y,numbers)

print(product)
```

```python
# 12. Difference Between Lambda and Normal Function
# Problem Statement: Write the same logic using a normal function and a lambda function

# A normal function is defined using def and can contain multiple lines. A lambda function is written in
# one line and is useful for short operations. Both examples return the same result

def add(a , b):
  return a + b

add_lambda = lambda a , b : a + b

print(add(58,75))
print(add_lambda(58,75))

```
