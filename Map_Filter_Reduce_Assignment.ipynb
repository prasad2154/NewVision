```python
# 1. Square Each Number Using map()
numbers = [54,75,43,74,12,98]

result = list(map(lambda x:x**2,numbers))
print(result)
```

```python
# 2. Convert Celsius to Fahrenheit Using map()
# Interview Question: Convert a list of Celsius temperatures into Fahrenheit using map().

celsius = [70,32,0,20,200]
fahrenheit = list(map(lambda x: (float(9)/5)*x + 32, celsius))
print(fahrenheit)
```

```python
# 3. Extract Names from List of Dictionaries

students = [
  {'name': 'Amit', 'marks': 80},
 {'name': 'Pranav', 'marks': 92},
 {'name': 'Rahul', 'marks': 76}
]

names = list(map(lambda x: x['name'], students))
print(names)
```

```python
# 4. Filter Positive Numbers
numbers = [-54,34,-54,98,12,-21]

result=list(filter(lambda x:x >0,numbers))
print(result)
```

```python
# 5. Filter Students Who Passed
students = [
  {'name': 'Amit', 'marks': 80},
 {'name': 'Pranav', 'marks': 92},
 {'name': 'Rahul', 'marks': 76},
 { 'name': 'Raj', 'marks': 35}
]

result = list(filter(lambda s:s['marks']>=40,students))
print(result)
```

```python
# 6: Filter all words that start with the letter P..

word = ["prasad","samiksha","divakar","pandu","pritam" ]
letter = list(filter(lambda x:x.startswith('p'),word))
print(letter)
```

```python
# 7. Find Sum Using reduce()
from functools import reduce
numbers = [54,75,43,74,12,98]

result = reduce(lambda x,y:x+y,numbers)
print(result)
```

```python
# 8. Find Minimum Using reduce()
numbers = [54,75,43,74,12,98]

result = reduce(lambda x,y:x if x<y else y,numbers)
print(result)
```

```python
# 9. Count Total Characters in Words Using map() and reduce(
from functools import reduce
words = ['AI', 'Python', 'Data']
lengths = list(map(lambda word: len(word), words))
total_chars = reduce(lambda x, y: x + y, lengths)
print(total_chars)
```

```python
# 10. Calculate Total Bill After Discount
from functools import reduce

prices = [1000, 500, 800]
discounted = list(map(lambda p: p - (p * 0.10), prices))
total_bill = reduce(lambda x, y: x + y, discounted)
print(total_bill)
```
