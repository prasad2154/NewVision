```python
# 3.reduce(function.iterable)

# reduce is present in functools module

from functools import reduce

k=[2,3,4,5,6]

reduce(lambda x,y:x+y,k)
# ((((2+3)+4)+5)+6)=20
# In reduce, we don't need typecasting as it gives the result
```

```python
reduce(lambda x,y:max(x,y),k)    # finding max
```

```python
# OS MODULE

# The os module allows Python to interact with the Operating System. It helps you work with files, folders (directories), paths, and environment variables.

import os
# to print current working directory
print(os.getcwd())

# os.getcwd() means Get Current Working Directory.
# It returns the folder where your Python program is currently running.
```

```python
# sys module
# The sys module provides information and functions related to the Python interpreter.

# It helps us:

# Check Python version
# Exit a program
# Read command-line arguments
# Get memory size

import sys
print(sys.version)
```

```python
# io Module (Input/Output)

# The IO module is used to read data from files and write data into files.

# Python usually uses open(), which internally works with the io module.

# Syntax
# with open("student.txt", "w") as file:

with open("student.txt", "w") as file:
    file.write("Hello Python")
```

```python
# array Module

# An array stores multiple values of the same data type.

# Unlike a list, arrays cannot store different data types.

# Syntax
# from array import array

from array import array

numbers =array("i",[10,20,40])
print(numbers)
```

```python
# collections Module
# The collections module provides special data structures that are more powerful than normal lists and dictionaries.

# Some commonly used classes are:

# Counter
# deque
# defaultdict
# namedtuple

# Syntax
# from collections import Counter
from collections import Counter
text ="banana"
# Counter automatically counts how many times each character present

print(Counter(text))
```

```python
# # counter module
# Counter is a class inside the collections module.

# It counts how many times each item appears.

# Syntax
# from collections import Counter

from collections import Counter

numbers=[1,2,3,4,3,4,5,6,1,3,2,3]
print(Counter( numbers))
```

```python
# # itertools Module
# # The itertools module provides tools to create efficient iterator
# It is mainly used for:

# Combinations
# Permutations
# Infinite counting
# Repeating values
# Syntax
# from itertools import combinations

from itertools import combinations
numbers=[1,2,3,4,5]
print(list(combinations(numbers,3)))
# It creates all possible groups of 3 letters without repeating the same pair.
```
