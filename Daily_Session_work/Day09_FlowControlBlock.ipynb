```python
# it is used to fetch the elements from collection ONE by ONE

nms=['vandana','sarita','prasad','pratik','rajiv']

for n in nms:
  print(n.upper())
```

```python
# if I want such names who endswith letter a
for n in nms:
  if n.endswith('a'):
    print(n.upper())
```

```python
# fetch nammes in which i present
for n in nms:
  if 'i' in n:
    print(n.title())
```

```python
# Fetch names whose length is less than or equal to 5
for n in nms:
  if len(n) <= 5:
    print(n)
```

```python
# fetch numbers divisible by 5 and 7
nums=[23,10,70,45,90,100,35,7,8,16]

for n in nums:
  if n % 5 == 0 and n % 7 == 0:
    print(n)

```

```python
# Assignment

# i want to fetch most frequent item
k = [1,1,1,2,2,2,3,4,5,4,4,]

max_count = 0

for i in set(k):
    if k.count(i) > max_count:
        max_count = k.count(i)

for i in set(k):
    if k.count(i) == max_count:
        print(i)
```
