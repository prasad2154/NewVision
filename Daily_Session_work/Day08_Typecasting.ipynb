```python
# assignment-- create a simple bill of grocery take 3-4 inputs and print the final bill

print("welcome to our grocery store")

# Product 1
product_name1 = input("enter product 1 name: ")
quantity1 = float(input(f"enter quantity for {product_name1}: "))
price1 = float(input(f"enter price per unit for {product_name1}: Rs"))
total_product1 = quantity1 * price1

# Product 2
product_name2 = input("enter product 2 name: ")
quantity2 = float(input(f"enter quantity for {product_name2}: "))
price2 = float(input(f"enter price per unit for {product_name2}: Rs"))
total_product2 = quantity2 * price2

# Product 3
product_name3 = input("enter product 3 name: ")
quantity3 = float(input(f"enter quantity for {product_name3}: "))
price3 = float(input(f"enter price per unit for {product_name3}: Rs"))
total_product3 = quantity3 * price3

# Calculate final bill
final_bill = total_product1 + total_product2 + total_product3

print("\n--- Your Bill ---")
print(f"{product_name1}: {quantity1} x Rs{price1} = Rs{total_product1:.2f}")
print(f"{product_name2}: {quantity2} x Rs{price2} = Rs{total_product2:.2f}")
print(f"{product_name3}: {quantity3} x Rs{price3} = Rs{total_product3:.2f}")
print(f"\nTotal amount of bill is Rs{final_bill:.2f}")
```
