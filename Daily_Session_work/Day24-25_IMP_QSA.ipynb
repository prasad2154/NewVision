**Take one string and count vowels and consonants**

```python
s ="Hello all welcome to Python Programming"
vo=0
co=0
s=s.lower()
# remove  space
s=s.replace(" ", "")
vowels ="aeiou"
for item in s:
  if item in vowels:
    vo+=1 #increment vowel
  else:
    co+=1 # increment for consonnats

print("vowels count:",vo)
print("consonants count:",co)

```

```python
# 1. Check whether a number is a Palindrome or not
s = input("Enter a string: ")

rev = ""

for ch in s:
    rev = ch + rev

if s == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# 2. Check whether a number is an armstrong number or not-------An Armstrong Number is a number that is equal to the
# sum of its digits, where each digit is raised to the power of the total number of digits.
num = int(input("Enter a number: "))


temp = num
count = len(str(num))
sum = 0

while temp > 0:
    digit = temp % 10
    sum = sum + digit ** count
    temp = temp // 10

if sum == num:
    print(num, "is an Armstrong Number")
else:
    print(num, "is Not an Armstrong Number")


```

```python

#3.  Find a factorial of given number (without recrsion)
num = int(input("Enter a number: "))

fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial =", fact)
```

```python
#5. Generate the Fibonacci Series
# Ex. 7 ---> 0 1 1 2 3 5 8
# The Fibonacci Series is a sequence of numbers where each number is the sum of the two previous numbers.

num = int(input("Enter the number of terms: "))

a = 0
b = 1

for i in range(num):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```
