# **Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
# we canperform differnt types of analysis
# Univariate--we are using single feature
# Bi-variate-- we are using 2 features
# Multi-Variate--More than 2 features(state vs population vs gender)
```

```python
days = ['mon','tue','wed','thur','fri','sat','sun']
rain =[120,30,25,20,80,20,140]
```

```python
# Bi-variate---line,scatter,barplot
```

```python
plt.plot(days,rain,"g--*")
plt.show()
```

```python
plt.plot(days,rain,'y',label="Day wise rainfall in mm")
plt.legend
plt.grid()
plt.title("Weekly RainFall")
plt.xlabel("Days mon-fri")
plt.ylabel("RainFall in mm")
plt.legend()
plt.show()
plt.savefig("weekly_rainfall.png",dpi=500)
```

```python
plt.bar(days,rain,color= ['r','y','m','b','k'],label='Daywise rainfall in mm')
plt.legend
plt.grid()
plt.title("Weekly RainFall")
plt.xlabel("Days mon-fri")
plt.ylabel("RainFall in mm")
plt.legend()
plt.show()
```
